## MIARY ZMIENNOŚCI

```python
def miary_zmiennosci(x):
    #odchylenie standardowe
    stdev = statistics.stdev(x)
    
    #odchylenie przeciętne
    odch_przec = stats.median_abs_deviation(x)
    
    #odchylenie ćwiartkowe
    Q = np.quantile(x, [0.25, 0.75])
    odch_ćw = (Q[1] - Q[0]) / 2
    
    #semiodchylenie std
    semi_std = np.sqrt(np.sum(x[x -np.mean(x)<0] ** 2) / (len(x)-1) )
    
    #semiodchylenie przeciętne
    semi_przec = np.sum(np.abs(x[x -np.mean(x)<0] - np.mean(x))) / (len(x)-1)
    
    data = [['Odchylenie standardowe', stdev], ['Odchylenie przeciętne', odch_przec], ['Odchylenie ćwiartkowe', odch_ćw],
           ['Semiodchylenie standardowe', semi_std], ['Semiodchylenie przeciętne', semi_przec]] 
    df = pd.DataFrame(data, columns=['Miara zmienności', 'Wartość'])
    
    return df
```

```python
miary_zmiennosci(hao)
```

```python
miary_zmiennosci(hao1)
```

```python
miary_zmiennosci(mate)
```

```python
miary_zmiennosci(mate1)
```

## KWANTYLE

```python
def kwantyle(x1, x2):
    kwantyle = np.arange(0,1,0.05)
    Q1 = np.quantile(x1, kwantyle)
    Q2 = np.quantile(x2, kwantyle)
    df = pd.DataFrame({'Kwantyl':kwantyle, 'Wartość - niestacjonarne':Q1, 'Wartość - stacjonarne':Q2})
    
    x = np.arange(0, 1, 0.01)
    q1 = np.quantile(x1, x)
    q2 = np.quantile(x2, x)
    sns.set()
    plt.plot(x, q1, color='teal')
    plt.title('Wykres kwantylowy - dane niestacjonarne')
    plt.xlabel('Kwantyl')
    plt.ylabel("Wartość")
    plt.show()
    
    plt.plot(x, q2, color='teal')
    plt.title('Wykres kwantylowy - dane stacjonarne')
    plt.xlabel('Kwantyl')
    plt.ylabel('Wartość')
    plt.show()
    
    return df
```

```python
kwantyle(hao, hao1)
```

```python
kwantyle(mate, mate1)
```
